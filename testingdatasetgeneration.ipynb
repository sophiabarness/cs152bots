```python
import ast 
from collections import Counter

def convert_to_list(string):
    try:
        return ast.literal_eval(string)
    except (ValueError, SyntaxError):
        return []
    
df['hashtags'] = df['hashtags'].apply(convert_to_list)

all_hashtags = [hashtag for sublist in df['hashtags'] for hashtag in sublist]

hashtag_counts = Counter(all_hashtags)
```

```python
# Get the 50 most common hashtags
most_common_hashtags = hashtag_counts.most_common(50)

# Print the 50 most popular hashtags
for hashtag, count in most_common_hashtags:
    print(f'{hashtag}: {count}')

hashtags = []

for hashtag, count in most_common_hashtags:
    hashtags.append(hashtag)
```

```python
import os 

# Initialize an empty list to store filtered DataFrames
filtered_dfs = []

def contains_desired_hashtags(entities, hashtag_list):
    if 'hashtags' in entities:
        hashtags_in_tweet = [ht['text'] for ht in entities['hashtags']]
        return any(ht in hashtag_list for ht in hashtags_in_tweet)
    return False

# Iterate over the filenames
for j in range(20):
    for i in range(60):  # Assuming you want to iterate over files 01/00/00.json to 01/00/15.json
        filename = f'01/{j:02}/{i:02}.json'  # Construct the filename
        if os.path.exists(filename):  # Check if the file exists
            # Read the JSON file into a DataFrame
            df = pd.read_json(filename, lines=True)
            
            # Perform the filtering operations
            df = df.dropna(subset=['text'])
            df = df.drop(columns=['created_at', 'id', 'id_str', 'source', 'truncated', 'in_reply_to_status_id'])
            df = df.drop(columns=['in_reply_to_status_id_str', 'in_reply_to_user_id', 'in_reply_to_user_id_str', 'in_reply_to_screen_name'])
            df = df.drop(columns=['geo', 'coordinates', 'place', 'contributors', 'is_quote_status', 'quote_count', 'reply_count', 'retweet_count'])
            df = df.drop(columns=['favorite_count', 'favorited', 'retweeted', 'filter_level', 'display_text_range', 'possibly_sensitive', 'extended_entities', 'extended_tweet', 'quoted_status_id', 'quoted_status_id_str', 'quoted_status', 'quoted_status_permalink', 'delete'])
            df = df[df['retweeted_status'].isna()]
            df = df[df['entities'].apply(lambda x: 'hashtags' in x and len(x['hashtags']) > 0)]
            
            # Filter the DataFrame based on the presence of desired hashtags
            df_filtered = df[df['entities'].apply(lambda x: contains_desired_hashtags(x, all_hashtags))]
            
            # Append the filtered DataFrame to the list
            filtered_dfs.append(df_filtered)

# Concatenate all filtered DataFrames into one DataFrame
combined_df = pd.concat(filtered_dfs, ignore_index=True)

# Display information about the combined DataFrame
print(combined_df.info())
```

```python
def extract_hashtags(entities):
    if 'hashtags' in entities:
        return [ht['text'] for ht in entities['hashtags']]
    else:
        return []

# Create a new column 'hashtags' containing lists of hashtags
df['hashtags'] = df['entities'].apply(extract_hashtags)
print(df.head())
```

```python
df = df.drop(columns=['entities'])
```

```python
def extract_location(user):
    if 'location' in user:
        return user['location']
    else:
        return None

# Create a new column 'hashtags' containing lists of hashtags
df['user_reported_location'] = df['user'].apply(extract_location)
print(df.head())
```

```python
def extract_profile(user):
    if 'description' in user:
        return user['description']
    else:
        return None

# Create a new column 'hashtags' containing lists of hashtags
df['user_profile_description'] = df['user'].apply(extract_profile)
print(df.head())
```
